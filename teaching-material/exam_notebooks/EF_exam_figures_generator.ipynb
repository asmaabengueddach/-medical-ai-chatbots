```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
file_path = "donnees_diabetiques.csv"
data = pd.read_csv(file_path)

# Affichage des premières lignes
data.head()

# Visualisation des niveaux de glucose par catégorie
graph = sns.boxplot(x="Categorie", y="Niveau_Glucose", data=data)
plt.title("Distribution des niveaux de glucose par catégorie")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Générer des données synthétiques pour les fréquences cardiaques
np.random.seed(42)
frequence_cardiaque = {
    'Patient': range(1, 101),
    'Avant_Traitement': np.random.normal(85, 10, 100),
    'Apres_Traitement': np.random.normal(75, 8, 100)
}
df_frequence_cardiaque = pd.DataFrame(frequence_cardiaque)
df_frequence_cardiaque.to_csv('frequence_cardiaque.csv', index=False)

# Générer des données synthétiques pour les niveaux de glucose
glucose = {
    'Patient': range(1, 201),
    'Glucose': np.concatenate([
        np.random.normal(110, 15, 150),  # Normaux
        np.random.normal(200, 20, 50)   # Hyperglycémie
    ])
}
df_glucose = pd.DataFrame(glucose)
df_glucose.to_csv('glucose.csv', index=False)

print("Fichiers générés : 'frequence_cardiaque.csv' et 'glucose.csv'")
```

```python
# Importer les bibliothèques
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données de fréquence cardiaque
df_frequence_cardiaque = pd.read_csv('frequence_cardiaque.csv')

# Visualisation
plt.figure(figsize=(10, 6))
plt.boxplot(
    [df_frequence_cardiaque['Avant_Traitement'], df_frequence_cardiaque['Apres_Traitement']],
    labels=['Avant Traitement', 'Après Traitement']
)
plt.title('Fréquence cardiaque moyenne avant et après traitement')
plt.ylabel('Fréquence cardiaque (bpm)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Importer les bibliothèques
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données de glucose
df_glucose = pd.read_csv('glucose.csv')

# Visualisation
plt.figure(figsize=(10, 6))
sns.histplot(df_glucose['Glucose'], bins=30, kde=True, color='blue')
plt.title('Distribution des niveaux de glucose des patients')
plt.xlabel('Niveau de glucose (mg/dL)')
plt.ylabel('Fréquence')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
