## Partie 1 : Introduction aux Graphiques de Base

### **1. Graphique en ligne (Line Plot)**

Le graphique en ligne est utilisé pour représenter des **tendances au fil du temps** ou des **changements continus**.

```python
import matplotlib.pyplot as plt

# Exemple de données
jours = ['Dimanche', 'Lundi', 'Mardi', 'Mercredi', 'Jeudi']
ventes = [200, 450, 300, 500, 650]

# Créer un graphique en ligne
plt.plot(jours, ventes, marker='o', linestyle='-', color='blue')
plt.title("Ventes journalières")
plt.xlabel("Jours")
plt.ylabel("Ventes (€)")
plt.grid(True)
plt.show()
```

### **2. Histogramme**

Un histogramme montre **la répartition des données** en regroupant les valeurs dans des intervalles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Génération de données fictives avec randn
data = np.random.randn(1000)

# Créer un histogramme
plt.hist(data, bins=20, color='green', alpha=0.7, edgecolor='black')
plt.title("Répartition des valeurs")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()
```

Ici, **bins=30** signifie que l'histogramme sera créé avec **30 intervalles**. Les données seront divisées en **30 groupes**, et l'histogramme affichera la fréquence des données dans chaque groupe.

**Rappel :**

- La plupart des valeurs générées par **randn** seront comprises entre -3 et +3.

### **3. Graphique à barres**

Utile pour **comparer des catégories distinctes**.

```python
categories = ['Catégorie A', 'Catégorie B', 'Catégorie C']
valeurs = [40, 55, 70]

# Créer un graphique à barres
plt.bar(categories, valeurs, color='purple')
plt.title("Comparaison des catégories")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.show()
```

### **4. Nuage de points (Scatter Plot)**

Un nuage de points est utilisé pour **examiner les relations entre deux variables numériques**.


```python
import matplotlib.pyplot as plt

# Données
population = [328, 1439, 1366, 1281, 377]
pib = [21, 14, 2, 10, 6]
noms_pays = ["USA", "Chine", "Inde", "Japon", "France"]

# Nuage de points
plt.figure(figsize=(12, 6))
plt.scatter(population, pib)

# Annotations simplifiées
for i, nom in enumerate(noms_pays):
    plt.annotate(nom, (population[i], pib[i]), ha='right', va='bottom')

# Labels et titre
plt.xlabel("Population")
plt.ylabel("PIB")
plt.title("Relation entre la population et le PIB")
plt.show()
```

### **Exemple : Répartition des patients par département médical**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Exemple de données : Répartition des patients par département médical
data = {
    'Département': ['Cardiologie', 'Neurologie', 'Pédiatrie', 'Dermatologie', 'Oncologie'],
    'Patients': [120, 90, 60, 80, 100]
}

df = pd.DataFrame(data)

# Créer un camembert
plt.figure(figsize=(8, 8))
plt.pie(
    df['Patients'],
    labels=df['Département'],
    autopct='%1.1f%%'
)
plt.title("Répartition des Patients par Département Médical")
plt.show()
```

- **`autopct='%1.1f%%'`** signifie :

  - Afficher les pourcentages sur chaque part du camembert avec une largeur minimale de 1, une décimale, et le symbole de pourcentage après le nombre.

  - Par exemple, si une part représente 25.345% du total, elle sera affichée comme 25.3%.

#### **Interprétation :**

Un graphique en camembert montrant les pourcentages de patients pour chaque département médical :
- **Cardiologie : 27.9%**  
- **Neurologie : 20.9%**  
- **Pédiatrie : 14.0%**  
- **Dermatologie : 18.6%**  
- **Oncologie : 23.3%**

### **5. Heatmap avec Seaborn**

Une heatmap visualise **les corrélations ou relations entre plusieurs variables**.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement du jeu de données "iris"
iris = sns.load_dataset('iris')

# Sélectionner uniquement les colonnes numériques
numeric_iris = iris.select_dtypes(include=['number'])

# Calculer la matrice de corrélation
correlation_matrix = numeric_iris.corr()

# Créer une heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Matrice de Corrélation - Iris Dataset")
plt.show()
```

###**Interprétation des Couleurs et des Valeurs**

1. **Couleurs:**

  - Les couleurs chaudes (comme le rouge) indiquent une corrélation positive. Plus la couleur est intense, plus la corrélation est forte.
  - Les couleurs froides (comme le bleu) indiquent une corrélation négative.
  - Les couleurs proches du blanc indiquent une corrélation faible ou nulle.
2. **Valeurs:**

- Les valeurs affichées dans chaque carré représentent le coefficient de corrélation, qui varie de -1 à +1.
  - +1 : Corrélation positive parfaite (les variables évoluent dans le même sens).
  - -1 : Corrélation négative parfaite (les variables évoluent en sens inverse).
  - 0 : Pas de corrélation linéaire.

**Exemple d'Interprétation :**

Dans ce graphique, le carré rouge foncé avec une valeur proche de +1 entre `petal_length` et `petal_width`, signifie qu'il existe une forte corrélation positive entre la longueur et la largeur des pétales. En d'autres termes, lorsque la longueur des pétales augmente, la largeur des pétales a également tendance à augmenter.

### **6. Boxplot (Boîte à Moustache) avec Seaborn**

Un **boxplot** est utilisé pour résumer **la distribution d’un ensemble de données numériques**. Il met en évidence la `médiane`, les `quartiles` et les `valeurs aberrantes`.

### **Exemple : Analyse des Salaires des Employés**

Nous utilisons un dataset fictif sur les salaires des employés pour visualiser la répartition des salaires par département à l’aide d’un boxplot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Department': ['HR', 'HR', 'HR', 'HR', 'IT', 'IT', 'IT', 'IT', 'Finance', 'Finance', 'Finance', 'Finance'],
    'Salary': [45000, 52000, 51000, 60000, 65000, 72000, 71000, 80000, 75000, 82000, 81000, 90000]
}
df = pd.DataFrame(data)

# # Création d'un boxplot avec Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x='Department', y='Salary', data=df, palette="Set2")
plt.title("Répartition des Salaires par Département")
plt.xlabel("Département")
plt.ylabel("Salaire")
plt.show()
```

- **Explication**
  - Chaque département (HR, IT, Finance) aura une boîte à moustache qui résume la distribution des salaires.
  - La médiane sera représentée par une ligne à l'intérieur de la boîte.
  - Les valeurs aberrantes seront affichées sous forme de points au-dessus ou en dessous des "moustaches".


### **7. Barplot avec Seaborn**

Un **barplot** (ou graphique à barres) est utilisé pour visualiser des données catégoriques et leurs valeurs correspondantes. Il est particulièrement utile pour :

- Comparer les moyennes ou les sommes de différentes catégories.
- Identifier des tendances ou des disparités entre des groupes.
- Présenter des informations agrégées, comme les moyennes ou les totaux, avec une mesure d’incertitude (écart-type, intervalle de confiance).

#### **Exemple de Barplot avec hue :**
Répartition des Patients par Sexe et Département Médical

- Comparer les patients masculins et féminins dans différents départements médicaux.
- Analyser les différences entre sous-groupes au sein de catégories principales.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Création d'un dataset fictif
data = {
    'Département': ['Cardiologie', 'Cardiologie', 'Neurologie', 'Neurologie', 'Orthopédie', 'Orthopédie'],
    'Sexe': ['Homme', 'Femme', 'Homme', 'Femme', 'Homme', 'Femme'],
    'Nombre de Patients': [50, 60, 70, 80, 40, 45]
}

df = pd.DataFrame(data)

# Création du barplot avec Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Département', y='Nombre de Patients', hue='Sexe', data=df, ci=None)

# Ajout d'informations et personnalisation
plt.title("Répartition des Patients par Sexe et Département Médical", fontsize=14)
plt.xlabel("Département Médical", fontsize=12)
plt.ylabel("Nombre de Patients", fontsize=12)
plt.legend(title="Sexe", loc="upper right")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Affichage du graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Création d'un dataset
data = {
    'Produit': ['Produit A', 'Produit B', 'Produit C', 'Produit A', 'Produit B', 'Produit C'],
    'Ventes': [120, 90, 150, 180, 140, 200],
    'Mois': ['Janvier', 'Janvier', 'Janvier', 'Février', 'Février', 'Février']
}
df = pd.DataFrame(data)

# Création du graphique en ligne
plt.figure(figsize=(10, 6))

# Liste unique des produits
produits = df['Produit'].unique()

# Tracer une ligne pour chaque produit
for produit in produits:
    produit_data = df[df['Produit'] == produit]
    plt.plot(produit_data['Mois'], produit_data['Ventes'], marker='o', label=produit)

# Ajout des titres et légendes
plt.title("Tendance des Ventes par Produit au Fil des Mois", fontsize=14)
plt.xlabel("Mois", fontsize=12)
plt.ylabel("Ventes (unités)", fontsize=12)
plt.legend(title="Produit")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Affichage du graphique
plt.show()
```
